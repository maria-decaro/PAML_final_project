```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
```

```python
df=pd.read_csv("diabetes_dataset.csv")

df.head()
```

```python
df=df[df['DIABETE4'].isin([1,3])]

X = df.drop(columns=["DIABETE4"]).values
y = df["DIABETE4"].values.flatten()

feature_cols = [col for col in df.columns if col != "DIABETE4"]
```

First we take the diabetes data and make sure to use the parsed down data from our previous notebook. Next, we create the training/test split. The team agreed to use an 80/20 split, which we used for all of our models. 

```python
#Separating the training and testing data
X, y = df.loc[:, ~df.columns.isin(['DIABETE4'])], df.loc[:, df.columns.isin(['DIABETE4'])]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=1)
```

```python
print('Training set size: {}'.format(len(X_train)))
print('Test set size: {}'.format(len(X_test)))
```

Undersampling the Majority Class

```python
class UndersamplingPreprocessor:
    def fit_resample(self, X_train, y_train):
        # Convert to numpy
        X_train = np.array(X_train)
        y_train = np.array(y_train).flatten()
        
        # Separate classes
        X_majority = X_train[y_train == 3]
        X_minority = X_train[y_train == 1]
        
        y_majority = y_train[y_train == 3]
        y_minority = y_train[y_train == 1]
        
        # Get minority size
        n_minority = len(X_minority)
        
        # Randomly sample from majority WITHOUT replacement
        indices = np.random.choice(len(X_majority), size=n_minority, replace=False)
        
        X_majority_downsampled = X_majority[indices]
        y_majority_downsampled = y_majority[indices]
        
        # Combine
        X_train_balanced = np.vstack((X_majority_downsampled, X_minority))
        y_train_balanced = np.hstack((y_majority_downsampled, y_minority))
        
        # Shuffle
        shuffle_idx = np.random.permutation(len(y_train_balanced))
        X_train_balanced = X_train_balanced[shuffle_idx]
        y_train_balanced = y_train_balanced[shuffle_idx]

        return X_train_balanced, y_train_balanced
    def save(self, X_train_balanced, y_train_balanced, path="data_undersampling.csv"):
        df = pd.DataFrame(X_train_balanced, columns=feature_cols)
        df["DIABETE4"] = y_train_balanced
        df.to_csv(path, index=False)
        print(f"Saved to {path} — shape: {df.shape}")
```

Oversampling the Majority Class

```python
class OversamplingPreprocessor:
    def fit_resample(self, X_train, y_train):
        # Convert to numpy
        X_train = np.array(X_train)
        y_train = np.array(y_train).flatten()

        # Separate classes
        X_majority = X_train[y_train == 3]
        X_minority = X_train[y_train == 1]

        y_majority = y_train[y_train == 3]
        y_minority = y_train[y_train == 1]

        # Get majority size
        n_majority = len(X_majority)

        # Randomly sample from minority WITH replacement
        indices = np.random.choice(len(X_minority), size=n_majority, replace=True)

        X_minority_upsampled = X_minority[indices]
        y_minority_upsampled = y_minority[indices]

        # Combine
        X_train_balanced = np.vstack((X_majority, X_minority_upsampled))
        y_train_balanced = np.hstack((y_majority, y_minority_upsampled))

        # Shuffle
        shuffle_idx = np.random.permutation(len(y_train_balanced))
        X_train_balanced = X_train_balanced[shuffle_idx]
        y_train_balanced = y_train_balanced[shuffle_idx]

        return X_train_balanced, y_train_balanced

    def save(self, X_train_balanced, y_train_balanced, path="data_oversampling.csv"):
        df = pd.DataFrame(X_train_balanced, columns=feature_cols)
        df["DIABETE4"] = y_train_balanced
        df.to_csv(path, index=False)
        print(f"Saved to {path} — shape: {df.shape}")
```

SMOTE Pre-processing

```python
class SMOTEPreprocessor:
    def __init__(self, random_state=42):
        self.smote = SMOTE(random_state=random_state)

    def fit_resample(self, X, y):
        return self.smote.fit_resample(X, y)

    def save(self, X, y, path="data_smote.csv"):
        df = pd.DataFrame(X, columns=feature_cols)
        df["DIABETE4"] = y
        df.to_csv(path, index=False)
        print(f"Saved to {path} — shape: {df.shape}")
```

Now, we ran all three of these methods and saved to a CSV

```python
under = UndersamplingPreprocessor()
X_under, y_under = under.fit_resample(X_train, y_train)
under.save(X_under, y_under)

over = OversamplingPreprocessor()
X_over, y_over = over.fit_resample(X_train, y_train)
over.save(X_over, y_over)

smote = SMOTEPreprocessor()
X_smote, y_smote = smote.fit_resample(X_train, y_train)
smote.save(X_smote, y_smote)
```
